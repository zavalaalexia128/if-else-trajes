## Variables
1. CT ---> Costo de traje
2. DE ---> Descuento que se obtendra
3. PF ---> Precio final del traje
## Formulas
1. $ CT > 2500$
2. $ DE = CT*0.15$
3. $ DE = CT*0.08$
4. $ PF = CT - DE $
## Diagrama de flujo LucidChart
![Diagrama de flujo](https://raw.githubusercontent.com/zavalaalexia128/if-else-trajes/refs/heads/main/DF%20if%20else%20trajes%20descuento.png)

```python
# Para cuando es verdadero (Si)
# Calculando el costo del traje
# datos de entrada  CT, DE, PF
# mostrar traje si es falso
# Si es verdadero mostrar No alcanza
CT = float(input("Ingrese el costo del traje: "))
if CT > 2500:
    DE = CT * 0.15
else:
    DE = CT * 0.08
PF = CT - DE
print("El costo del traje es: ", PF)
print("El descuento es: ", DE)
print("Programa realizado por Zavala Medina NL = 54")

```
